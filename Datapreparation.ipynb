```python
import pandas as pd
import os
import json
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
daftar_file = [
    "LBCurrent Interest Questions.csv",
    "LBCurrent Tech Questions.csv",
    "LBLearning Path Answer.csv",
    "LBSkill Keywords.csv",
    "LBStudent Progress.csv",
    "LPCM-CL.csv",
    "LPCM-Course.csv",
    "LPCM-Learning Path.csv",
    "LPCM-LpCourse.csv",
    "LPCM-Tutorials.csv",
]

print(f"Total ada {len(daftar_file)} file yang akan diproses.")
```

```python
def bersihkan_teks(teks):
    if pd.isna(teks):
        return ""
    teks = str(teks).strip()
    return teks
```

```python
database_gabungan = []

for nama_file in daftar_file:
    print(f"Sedang memproses: {nama_file}...", end=" ")

    try:
        df = pd.read_csv(nama_file)
        jumlah_awal = len(df)

        # --- PROSES TRANSFORMASI ---
        for index, row in df.iterrows():
            konten_teks = f"Sumber Data: {nama_file}\n"


            for kolom, nilai in row.items():
                nilai_bersih = bersihkan_teks(nilai)
                if nilai_bersih:
                    konten_teks += f"- {kolom}: {nilai_bersih}\n"

            database_gabungan.append({
                "source": nama_file,
                "content": konten_teks
            })

        print(f" Sukses! ({jumlah_awal} baris)")

    except FileNotFoundError:
        print(f" Gagal: File tidak ditemukan.")
    except Exception as e:
        print(f" Error: {e}")

print(f"\nTotal data terkumpul: {len(database_gabungan)} baris data.")
```

```python
# Lihat 3 data pertama
print("--- CONTOH DATA YANG SUDAH DIBERSIHKAN ---\n")
for i in range(3):
    if i < len(database_gabungan):
        print(f"DATA KE-{i+1}:")
        print(database_gabungan[i]['content'])
        print("-" * 30)
```

```python
nama_file_output = "knowledge_base.json"

with open(nama_file_output, "w", encoding='utf-8') as f:
    json.dump(database_gabungan, f, ensure_ascii=False, indent=4)

print(f"SELESAI! Data berhasil disimpan ke '{nama_file_output}'")
print(f"Sekarang kamu bisa update main.py kamu untuk membaca file ini saja.")
```
